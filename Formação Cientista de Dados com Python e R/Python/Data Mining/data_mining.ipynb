```python
import nltk
nltk.download('stopwords')
import string
import matplotlib.pyplot as plt
from nltk.corpus import PlaintextCorpusReader, stopwords
from matplotlib.colors import ListedColormap
from wordcloud import WordCloud
```

```python
corpus = PlaintextCorpusReader('../../Arquivos/Arquivos', '.*', encoding='ISO-8859-1')
```

```python
#Leitura
arquivus = corpus.fileids()
arquivus[0]
```

```python
# zero a 10
arquivus[0:10]
```

```python
#imprime todos os nomes
for a in arquivus:
    print(a)
```

```python
#Acesso aoo texto do primeiro arquivo
texto = corpus.raw('1.txt')
texto
```

```python
# A todos os textos de todos os arquivos
todo_texto = corpus.raw()
#todo_texto
```

```python
#obtendo as palavras do corpus e visualizando a quantidade
palavras = corpus.words()
palavras[132]
```

```python
#quantidade
len(palavras)
```

```python
#Usando o NLTK temos as stop word em inglês
stops = stopwords.words('english')
stops
```

```python
#definindo cores pra nuvem
mapa_cores = ListedColormap(['orange','blue','magenta','red','green'])
# criando a nuvem
nuvem = WordCloud(background_color='white',
                  colormap=mapa_cores,
                  stopwords= stops,
                  max_words=100)
#criando e visualizando a nuvem
nuvem.generate(todo_texto)
plt.imshow(nuvem)
```

```python
#Criando nova lista
palavras_semstop = [p for p in palavras if p not in stops]
len(palavras_semstop)
```

```python
#removendo a pontuação
palavras_sempontuacao = [p for p in palavras_semstop if p not in string.punctuation]
len(palavras_sempontuacao)
```

```python
#calculo da frequencia
frequencia = nltk.FreqDist(palavras_sempontuacao)
frequencia
```

```python
#mais comuns
mais_comuns = frequencia.most_common(100)
mais_comuns
```
