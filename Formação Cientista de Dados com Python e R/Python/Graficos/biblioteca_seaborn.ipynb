Grafico de Dispers√£o

```python
base3 = pd.read_csv('co2.csv')
base3.head()
```

```python
sea.scatterplot(base3.conc , base3.uptake, hue = base3.Type)
```

```python
q = base3.loc[base3['Type'] == 'Quebec']
m = base3.loc[base3['Type'] == 'Mississippi']
```

```python
plt.figure()
plt.subplot(1,2,1)
sea.scatterplot(q.conc, q.uptake).set_title('Quebec')
plt.subplot(1,2,2)
sea.scatterplot(m.conc, m.uptake).set_title('Mississippi')
plt.tight_layout()
```

```python
#Refrigerado/n√£o refrigerado
ch = base3.loc[base3['Treatment'] == 'chilled']
nc = base3.loc[base3['Treatment'] == 'nonchilled']
```

```python
plt.figure()
plt.subplot(2,2,1)
sea.scatterplot(ch.conc, ch.uptake).set_title('chilled')
plt.subplot(1,2,2)
sea.scatterplot(nc.conc, nc.uptake).set_title('nonchilled')
plt.tight_layout()
```
